```python
frases = []
hash_dado_sha256 = []
hash_dado_md5 = []
hash_calc_sha256 = []
hash_calc_md5 = []
sha_result = []
md5_result = []

for i in hash_dict.keys():
    SHA256 = hashlib.sha256(i.encode('utf-8')).hexdigest()
    MD5 = hashlib.md5(i.encode('utf-8')).hexdigest()

    bool_sha = hash_dict[i][0] == SHA256
    bool_md5 = hash_dict[i][1] == MD5
    
    frases.append(i)
    hash_dado_sha256.append(hash_dict[i][0]) 
    hash_dado_md5.append(hash_dict[i][1])     
    hash_calc_sha256.append(SHA256)
    hash_calc_md5.append(MD5)
    
    sha_result.append(bool_sha)
    md5_result.append(bool_md5)

df = pd.DataFrame({
    "Frase": frases,
    "SHA256 esperado": hash_dado_sha256,
    "SHA256 calculado": hash_calc_sha256,
    "SHA256 correto?": sha_result,
    "MD5 esperado": hash_dado_md5,
    "MD5 calculado": hash_calc_md5,
    "MD5 correto?": md5_result
})

df.to_excel("Lab_02.xlsx", index=False)
```
